# Make Your Own Movie
This notebook gets you started making your own movie, playing with an historically important dataset. We'll use the 25 pulsation periods and apparent magnitudes going into the first "Leavitt Law" period-luminosity relationship of Cepheid variable stars ([Leavitt & Pickering 1912](https://ui.adsabs.harvard.edu/abs/1912HarCi.173....1L/abstract)). This relationship is named for [Henrietta Swan Leavitt](https://en.wikipedia.org/wiki/Henrietta_Swan_Leavitt), the astronomer who discovered it. It was these variable stars that enabled astronomers to discover that the Universe is expanding.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Let's load and tidy the data.

```python
# the table is a default csv format, so it's easy to read
table = pd.read_csv('leavitt-periods.csv')
```

```python
# calculate the quantities we want 
log_period = np.log10(table['period'])
average_magnitude = 0.5*(table['min_magnitude'] + table['max_magnitude'])
```

Let's plot the data with a line fitted to it.

```python
N = 25

# fit a line to the first N data points
slope, intercept = np.polyfit(log_period[:N], average_magnitude[:N], 1)

# plot that model fit
x = np.linspace(0, 2.2)
plt.plot(x, slope*x + intercept)
        
# plot the data on top
plt.scatter(log_period[:N], average_magnitude[:N], color='black');
plt.xlabel('$log_{10}$(period/days)')
plt.ylabel('Average Magnitude');
plt.title(f'The Leavitt Law from {N} Cepheid Variable Stars');
```
